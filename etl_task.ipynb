#transform

```python
df = df.drop_duplicates()
```

```python
#because I dont have so much data , i will not drop the missing values , instead i will fill them with not available
cols = ['instructor_diploma', 'DiplomaDate', 'StudentBirthDate']
df[cols] = df[cols].fillna('not available')
```

```python
df.isna().sum().sum()
```

```python
df.head()
```

###transform  from object type to date type  

```python
df['StudentBirthDate'] = pd.to_datetime(df['StudentBirthDate'], errors='coerce')
df['SubscriptionStartDate'] = pd.to_datetime(df['SubscriptionStartDate'], errors='coerce')
df['SubscriptionEndDate'] = pd.to_datetime(df['SubscriptionEndDate'], errors='coerce')
```
